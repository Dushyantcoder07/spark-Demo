```python
from kafka import KafkaProducer
import json

# Fixed the spelling of bootstrap_servers here:
producer = KafkaProducer(
    bootstrap_servers="kafka:9092",
    value_serializer=lambda x: json.dumps(x).encode("utf-8")    
)

for i in range(10):
    producer.send(
        "demo-topic",
        {"id": i, "name": f"student-{i}"}
    )
producer.flush()
print("done")
```
